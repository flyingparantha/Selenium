```python
%%time #To check CODE-Run Time

#Because we Always Start with IMPORT
import os
import selenium
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
import time
import io 

#Function to retry till the Web-Page Loads 5 Times
def load_page(func, arg, browser):
    load_retry = 0
    while 1:
        if load_retry > 5:
            return False
        try:
            func(arg)
            print('load success!')
            return True
        except TimeoutException:
            load_retry += 1
            print('retrying')

            try:
                browser.execute_script("window.stop();")
            except TimeoutException:
                sleep(2)
                continue
                
#Script to access Download-Progress from shadowRoot folder of CHROME              
def get_download_progress():
    progress = browser.execute_script('''
    
    var tag = document.querySelector('downloads-manager').shadowRoot;
    var intag = tag.querySelector('downloads-item').shadowRoot;
    var progress_tag = intag.getElementById('progress');
    var progress = null;
    if(progress_tag) {
        progress = progress_tag.value;
    }
    return progress;
    
    ''')
    return progress

#Let's Load THE DRIVER
browser = webdriver.Chrome(executable_path=r"path_to_driver")
browser.set_page_load_timeout(15)
browser.maximize_window()
url = 'The_url'
load_page(browser.get, url, browser)
time.sleep(8)
browser.set_page_load_timeout(50)

browser.find_element_by_id("username").send_keys("usr")
browser.find_element_by_id("password").send_keys("pwd")
browser.find_element_by_id("Submit_button").submit()
print('login success!')
time.sleep(8)
browser.find_element_by_xpath('//a[contains(@href,"path")]').click()
time.sleep(8)
browser.find_element_by_xpath("path").click()
time.sleep(5)
browser.get('chrome://downloads/')

To Get Download Progress
progress = get_download_progress()
print('File Downloading.......')
while progress <100: 
    print("#", end =" ")
    time.sleep(1.5)
    progress = get_download_progress()
else:
    print("100%")
    print("File Downloaded!")      
    time.sleep(8)
    browser.close()
    browser.quit()

#Let's Wrap it Here by Closing and Quitting THE BROWSER
```
